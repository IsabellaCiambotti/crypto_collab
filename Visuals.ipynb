```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tbt = pd.read_parquet('tBTCUSD_trades.parquet', engine='fastparquet')
teo = pd.read_parquet('tEOSUSD_trades.parquet', engine='fastparquet')
tet = pd.read_parquet('tETHUSD_trades.parquet', engine='fastparquet')
```

```python
plots = sns.jointplot(x = tbt["MTS"], y = tbt["PRICE"], data=tbt, kind='scatter', color='goldenrod')
#axes = plots.axes
plt.title("BTC to USD")
```

```python
plots2 = sns.jointplot(x = teo["MTS"], y = teo["PRICE"], data=teo, kind='scatter', color='purple')
#axes = plots.axes
plt.title("EOS to USD")
```

```python
plots3 = sns.jointplot(x = tet["MTS"], y = tet["PRICE"], data=tet, kind='scatter', color='red')
#axes = plots.axes
plt.title("ETH to USD")
```
